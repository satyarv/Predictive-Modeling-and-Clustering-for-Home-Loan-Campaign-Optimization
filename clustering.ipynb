```python
import pandas as pd

data=pd.read_csv("/content/Population_Data.csv")
data.info()
```

```python
data.head()
```

```python
numeric = ["Indians", "Foreigners",
           "Indian_Male", "Indian_Female",
           "Foreigners_Female",
           "Foreigners_Male", "Total Population"]
```

```python
def cleaner(z):
    return z.replace(',', '')
```

```python
for i in data[numeric]:
    data[i] = data[i].apply(cleaner)

data.head()

```

```python
data[numeric] = data[numeric].apply(pd.to_numeric)
data.info()
```

```python
data[['Indians', 'Foreigners']].sum().sum() - data['Total Population'].sum()
```

```python
data[['Indian_Male', 'Indian_Female', 'Foreigners_Male', 'Foreigners_Female']].sum().sum() - data['Total Population'].sum()
```

```python
MF_sum = data['Indian_Male'] + data['Indian_Female'] + data["Foreigners_Male"] + data['Foreigners_Female']
data['other'] = data['Total Population'] - MF_sum
data.head()
```

```python
data['Region'].nunique(), data['Office Location Id'].nunique()
```

```python
data1 = data.drop(columns=['Region', 'Office Location Id', 'Total Population'])
data1.head()
```

```python
from sklearn.preprocessing import Normalizer

norm = Normalizer()                         # Create a Normalizer object
columns = data1.columns                     # Store column names
data1 = norm.fit_transform(data1)           # Normalize the data
data1 = pd.DataFrame(data1, columns=columns) # Convert back to DataFrame
data1.head()                                # Show first 5 rows
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)     # Initialize KMeans with 2 clusters
kmeans.fit(data1)                 # Fit the model on normalized data
pred = kmeans.predict(data1)     # Predict the cluster for each data point

pred, len(pred)                  # Output: cluster labels & number of rows
```

```python
kmeans.inertia_
```

```python
kmeans.score(data1)
```

```python
SSE = []
for cluster in range(1, 10):
    kmeans = KMeans(n_clusters=cluster, random_state=42)
    kmeans.fit(data1)
    SSE.append(kmeans.inertia_)
```

```python
frame = pd.DataFrame({'Cluster': range(1, 10), 'SSE': SSE})
frame
```

```python
import matplotlib.pyplot as plt

plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('inertia')
plt.title('Elbow Method for Optimal')
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data1)
```

```python
pred = kmeans.predict(data1)
```

```python
data1['cluster'] = pred
data1
```

```python
def seg(str_x, str_y, clusters):
    x = []
    y = []

    for i in range(clusters):
        x.append(data1[str_x][data1['cluster'] == i])
        y.append(data1[str_y][data1['cluster'] == i])

    return x, y
```

```python
def plot_clusters(str_x, str_y, clusters):
    plt.figure(figsize=(5, 5), dpi=120)

    x, y = seg(str_x, str_y, clusters)

    for i in range(clusters):
        plt.scatter(x[i], y[i], label='cluster{}'.format(i))

    plt.xlabel(str_x)
    plt.ylabel(str_y)
    plt.title(str_x + " vs " + str_y)
    plt.legend()
```

```python
plot_clusters('Indians', 'Foreigners', 3)
```
